```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_excel('Telco-Customer-Churn.xlsx')
```

```python
data
```

```python
data.isnull().sum()
```

### 1.Compare churn count with respect to gender

```python
plt.figure(figsize=(6,5))
sns.countplot('gender',hue='Churn',data=data)
```

### 2.Find out how many female senior citizens there in the dataset

```python
count1 = data[(data['gender']=='Female') & (data['SeniorCitizen']==1)]['gender'].count()
data1=data[(data['gender']=='Female') & (data['SeniorCitizen']==1)]
print("count of female Senior Citizens :",count1)
plt.title('Count of female senior citizens')
sns.countplot('gender',data=data1)
```

### 3. Compare 'tenure' with 'Total Charges'


```python
plt.figure(figsize=(15,10))
sns.scatterplot(data["tenure"],data["TotalCharges"]).set(title="Scatterplot of tenure and totalcharges")
```

### 4. Find out which contract is preferred by the senior citizen.


```python
data2=data[(data['SeniorCitizen']==1)]
plt.title("Senior Citizen Prefered Contract")
sns.countplot('Contract',data=data2)
```
